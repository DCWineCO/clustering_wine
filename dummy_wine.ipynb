```python
def concat_dataframe():
    '''
    This function adds a column to each dataframe that indicates red or white wine. 
    The two dataframes are then vertically concatenated into one dataframe keeping 
    original columns and display values vertically. 
    '''
    #added white column to distinct wine type
    df_white['white'] = 'white'
    #added red column to distinct wine type
    df_red['red'] = 'red'
    #two data frames are similar in column names, vertically concated on the rows 
    #to keep the column names and express all values vertically. 
    vertical_concat = pd.concat([df_white, df_red], axis=0)
    df = vertical_concat
    return df
```
